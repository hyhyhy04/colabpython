```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 파일 불러오기 (업로드한 후 실행)
df_culture = pd.read_csv('culture_participation.csv', encoding='cp949')
df_life = pd.read_csv('life_satisfaction.csv', encoding='cp949')
```

```python
# 2. life_satisfaction 전처리
# '행정구역별(1)'이 지역 이름인 열
df_life = df_life.rename(columns={'행정구역별(1)': '지역'})
df_life = df_life[df_life['지역'] != '행정구역별(1)']  # 첫 행 제거

# 매우만족 / 만족만 추출해서 melt
years = ['2011', '2013', '2015', '2017', '2019', '2021', '2023']
very_satisfied_cols = years
satisfied_cols = [f"{year}.1" for year in years]

df_very = df_life[['지역'] + very_satisfied_cols].copy()
df_very = df_very.melt(id_vars='지역', var_name='년도', value_name='매우만족')

df_satis = df_life[['지역'] + satisfied_cols].copy()
df_satis.columns = ['지역'] + years  # 열 이름 정리
df_satis = df_satis.melt(id_vars='지역', var_name='년도', value_name='만족')

# 보통 컬럼: 보통은 f"{year}.2" 형식
neutral_cols = [f"{year}.2" for year in years]

df_neutral = df_life[['지역'] + neutral_cols].copy()
df_neutral.columns = ['지역'] + years  # 열 이름 통일
df_neutral = df_neutral.melt(id_vars='지역', var_name='년도', value_name='보통')


# 기존: 매우만족과 만족만 병합
# df_life_long = pd.merge(df_very, df_satis, on=['지역', '년도'])

# 수정: 매우만족 + 만족 + 보통 병합
df_life_long = pd.merge(df_very, df_satis, on=['지역', '년도'])
df_life_long = pd.merge(df_life_long, df_neutral, on=['지역', '년도'])


# 3. culture_participation 전처리
df_culture = df_culture.rename(columns={'행정구역별(1)': '지역'})
df_culture = df_culture[df_culture['지역'] != '행정구역별(1)']

df_culture_melted = df_culture.melt(id_vars='지역', var_name='년도', value_name='문화예술관람률')

# 년도 형식 맞추기 (문자형으로)
df_culture_melted['년도'] = df_culture_melted['년도'].astype(str)
df_life_long['년도'] = df_life_long['년도'].astype(str)

# 4. 두 데이터 병합
df_merged = pd.merge(df_life_long, df_culture_melted, on=['지역', '년도'])

# 5. 숫자형 변환
for col in ['매우만족', '만족', '보통', '문화예술관람률']:
    df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')

# 6. 분석 & 시각화 함수
def plot_relation(x, y, title):
    plt.figure(figsize=(8,6))
    sns.regplot(data=df_merged, x=x, y=y, scatter_kws={'alpha':0.6})
    plt.title(title)
    plt.xlabel('문화예술 관람률 (%)')
    plt.ylabel(f'{y} (%)')
    plt.grid(True)
    plt.show()

# 7. 시각화 실행
plot_relation('문화예술관람률', '매우만족', '문화예술 관람률 vs 매우만족도 (전체 연도)')
plot_relation('문화예술관람률', '만족', '문화예술 관람률 vs 만족도 (전체 연도)')
plot_relation('문화예술관람률', '보통', '문화예술 관람률 vs 보통 만족도 (전체 연도)')
```

```python
# 2. 문화예술 관람률 데이터: 시도별, 연도 컬럼만 추출
culture = culture[['행정구역별(1)', '2021', '2019', '2017', '2015', '2013', '2011']].copy()
culture = culture.set_index('행정구역별(1)').astype(float)
culture['관람률_평균'] = culture.mean(axis=1)
culture = culture[['관람률_평균']]

# 3. 삶의 만족도: 매우만족 & 만족 따로 추출
# 매우만족
life_very = life[(life['특성별(1)'] == '매우만족') & (life['특성별(2)'] == '계')].copy()
life_very = life_very[['행정구역별(1)', '2021', '2019', '2017', '2015', '2013', '2011']]
life_very.columns = ['시도', '2021', '2019', '2017', '2015', '2013', '2011']
life_very = life_very.set_index('시도').astype(float)
life_very['매우만족_평균'] = life_very.mean(axis=1)

# 만족
life_satis = life[(life['특성별(1)'] == '만족') & (life['특성별(2)'] == '계')].copy()
life_satis = life_satis[['행정구역별(1)', '2021', '2019', '2017', '2015', '2013', '2011']]
life_satis.columns = ['시도', '2021', '2019', '2017', '2015', '2013', '2011']
life_satis = life_satis.set_index('시도').astype(float)
life_satis['만족_평균'] = life_satis.mean(axis=1)

# 4. 병합
merged_very = pd.merge(culture, life_very[['매우만족_평균']], left_index=True, right_index=True)
merged_satis = pd.merge(culture, life_satis[['만족_평균']], left_index=True, right_index=True)

# 5. 시각화
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 5))

# 매우만족
plt.subplot(1, 2, 1)
sns.regplot(data=merged_very, x='관람률_평균', y='매우만족_평균')
plt.title('문화예술 관람률 vs 매우만족 평균')
plt.xlabel('문화예술 관람률 평균(%)')
plt.ylabel('매우만족 평균(%)')

# 만족
plt.subplot(1, 2, 2)
sns.regplot(data=merged_satis, x='관람률_평균', y='만족_평균')
plt.title('문화예술 관람률 vs 만족 평균')
plt.xlabel('문화예술 관람률 평균(%)')
plt.ylabel('만족 평균(%)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 파일 불러오기
df_culture = pd.read_csv('culture_participation.csv', encoding='cp949')
df_life = pd.read_csv('life_satisfaction.csv', encoding='cp949')
# 1. life_satisfaction 전처리
df_life = df_life.rename(columns={'행정구역별(1)': '지역'})
df_life = df_life[df_life['지역'] != '행정구역별(1)']  # 첫 행 제거

# 연도 설정
years = ['2011', '2013', '2015', '2017', '2019', '2021', '2023']

# 매우만족, 만족, 보통 열 나누기
very_satisfied_cols = years
satisfied_cols = [f"{year}.1" for year in years]
normal_cols = [f"{year}.2" for year in years]

# 각각 melt 처리
df_very = df_life[['지역'] + very_satisfied_cols].copy()
df_very = df_very.melt(id_vars='지역', var_name='년도', value_name='매우만족')

df_satis = df_life[['지역'] + satisfied_cols].copy()
df_satis.columns = ['지역'] + years
df_satis = df_satis.melt(id_vars='지역', var_name='년도', value_name='만족')

df_normal = df_life[['지역'] + normal_cols].copy()
df_normal.columns = ['지역'] + years
df_normal = df_normal.melt(id_vars='지역', var_name='년도', value_name='보통')

# 병합
df_life_long = df_very.merge(df_satis, on=['지역', '년도']).merge(df_normal, on=['지역', '년도'])

# 2. culture_participation 전처리
df_culture = df_culture.rename(columns={'행정구역별(1)': '지역'})
df_culture = df_culture[df_culture['지역'] != '행정구역별(1)']
df_culture_melted = df_culture.melt(id_vars='지역', var_name='년도', value_name='문화예술관람률')

# 연도 형식 맞추기
df_culture_melted['년도'] = df_culture_melted['년도'].astype(str)
df_life_long['년도'] = df_life_long['년도'].astype(str)

# 병합
df_merged = pd.merge(df_life_long, df_culture_melted, on=['지역', '년도'])

# 숫자형 변환
for col in ['매우만족', '만족', '보통', '문화예술관람률']:
    df_merged[col] = pd.to_numeric(df_merged[col], errors='coerce')

# ✅ 새로운 계산: (만족 * 1.5 + 보통)
df_merged['만족_보통_조합'] = df_merged['만족'] * 1.5 + df_merged['보통']

# 시각화 함수
def plot_relation(x, y, title):
    plt.figure(figsize=(8,6))
    sns.regplot(data=df_merged, x=x, y=y, scatter_kws={'alpha':0.6})
    plt.title(title)
    plt.xlabel(f'{x} (%)')
    plt.ylabel(f'{y} (%)')
    plt.grid(True)
    plt.show()

# ✅ 시각화 실행: 삶의 만족도 vs (만족 * 1.5 + 보통)
plot_relation('문화예술관람률', '만족_보통_조합', '문화예술 관람률 vs 만족*1.5 + 보통 (전체 연도)')
```
